# load file

```python
from scipy.io import loadmat
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = loadmat("Shoe25OpenSim_python.mat")
print(data.keys())

angles = data['ANGLES_TABLE'][0,0]
print(angles.dtype.names)

trial = angles["x25_Curve_1_mat"]
print(type(trial))

print(trial.shape)
```

```python
import matplotlib.pyplot as plt

time = trial[:,0]
var1 = trial[:,1]

plt.plot(time, var1)
plt.xlabel("Time (s)")
plt.ylabel("Angle")
plt.title("Joint angle")
plt.show()
```

```python
data_labels = loadmat("angles_with_labels.mat")

raw_labels = data_labels["angle_labels"]

labels = [str(x[0]) for x in raw_labels[0]]

print(labels)
```

```python
angle_dict = {}

for n in range(len(labels)):
    angle_dict[labels[n]] = trial[:,n]
```

```python
time = angle_dict["time"]

ankle_angle = angle_dict['ankle_angle_r']

plt.plot(time, ankle_angle)
plt.xlabel("Time (s)")
plt.ylabel("Angle")
plt.title("Ankle angle R")
plt.show()

# mtp_angle = angle_dict['mtp_angle_r']

# plt.plot(time, mtp_angle)
# plt.xlabel("Time (s)")
# plt.ylabel("Angle")
# plt.title("Sub T angle R")
# plt.show()

hip_angle_R = angle_dict['hip_flexion_r']

plt.plot(time, hip_angle_R)
plt.xlabel("Time (s)")
plt.ylabel("Angle")
plt.title("Hip angle R")
plt.show()

hip_angle_L = angle_dict['hip_flexion_l']
plt.plot(time, hip_angle_L)
plt.xlabel("Time (s)")
plt.ylabel("Angle")
plt.title("Hip angle L")
plt.show()
```
